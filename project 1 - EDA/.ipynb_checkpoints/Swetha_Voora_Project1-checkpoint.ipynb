

# Finding Outliers

- Approach 1 : Using Z-score
- Approach 2 : Using Box-plot

## Finding Outliers using z-score

```python
# Specify the columns for which you want to calculate Z-scores and identify outliers
columns_to_check = ['TotalCharges', 'MonthlyCharges', 'tenure']

# Initialize an empty DataFrame to store outliers
outliers = pd.DataFrame()

# Calculate Z-scores and identify outliers for each column
for column in columns_to_check:
    z_scores = stats.zscore(data[column]) # Calculate the Z-score for the 'values' column
    threshold = 3 # Define a threshold for considering a point as an outlier (e.g., Z-score > 3)
    column_outliers = data[abs(z_scores) > threshold] # Identify outliers based on the threshold
    outliers = pd.concat([outliers,column_outliers])

print("Outliers:")
print(outliers)
```


## This above graph shows positive correlation betwen tenure and customer satisfaction.

From the above two cells, It is evident that:

```python
sns.lmplot(data=data, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

## The above plot shows Positive Correlation between Monthly charges and total charges.

```python
# KDE(Kernel Density Estimation) plot is a way to visualize the distribution of data in a smooth and continuous manner. 
Mth = sns.kdeplot(data.MonthlyCharges[(data["Churn"] == 'No') ],
                color="Red", fill = True)
Mth = sns.kdeplot(data.MonthlyCharges[(data["Churn"] == 'Yes') ],
                ax =Mth, color="Blue", fill= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

### Here it is evident that when the monthly charges are raising, the customers churn is raising too.
### It implies that the monthlycharges and the customer satisfaction are negatively correlated.

```python
Tot = sns.kdeplot(data.TotalCharges[(data["Churn"] == 'No') ],
                color="Red", fill = True)
Tot = sns.kdeplot(data.TotalCharges[(data["Churn"] == 'Yes') ],
                ax =Tot, color="Blue", fill= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

### - we observe that the churn rate is higher when the total charges are lower.
- The customers stayed loyal as their total charges have reached beyond 8000 usd
