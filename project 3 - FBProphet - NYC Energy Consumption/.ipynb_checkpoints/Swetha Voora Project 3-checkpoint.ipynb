## Creating a Daily Dataset...i.e., each record with a single date.

### Since the current dataset gives us the consumption over a period of days. We will create a record for each date within the period(# days) and then sum the consumption of energy in all areas in a single day

```python
# Calculate daily consumption
data_cleaned['Daily Consumption'] = data_cleaned['Consumption (KWH)'] / data_cleaned['# days']

# Create a daily date range for each record
data_cleaned['Date Range'] = data_cleaned.apply(lambda row: pd.date_range(row['Service Start Date'], row['Service End Date']), axis=1)

# Explode the DataFrame to have a row for each day
daily_data = data_cleaned.explode('Date Range')

# Rename and select relevant columns
daily_data = daily_data.rename(columns={'Date Range': 'Each Date'})[['Each Date', 'Daily Consumption']]

# Display the new DataFrame
print(daily_data)
```

## Daily Dataset

```python
from sklearn.metrics import mean_absolute_error
# Assuming 'train_365' and 'test_365' are your dataframes
periods = [100, 200, 365]
fig, axes = plt.subplots(3, 1, figsize=(10, 15))  # Adjust figure size as needed

for i, period in enumerate(periods):
    # Prepare the data for Prophet
    prophet_data = daily_train.rename(columns={'Service Start Date': 'ds', 'Consumption (KWH)': 'y'})

    prophet_model = Prophet()
    # Initialize and configure the Prophet model
#     prophet_model = Prophet(seasonality_mode='additive',
#                             yearly_seasonality=False,
#                             weekly_seasonality=False,
#                             daily_seasonality=False
#                             ).add_seasonality(name='monthly', period=30, fourier_order=4
#                             ).add_seasonality(name='yearly', period=365, fourier_order=4)
    
    prophet_model.fit(prophet_data)

    # Make future dataframe and predict
    future = prophet_model.make_future_dataframe(periods=period, include_history=False, freq='D')
    forecast = prophet_model.predict(future)

    # Aligning the forecast with the test data
    forecast_aligned = forecast[forecast['ds'].isin(daily_test['Service Start Date'])]
    test_aligned = daily_test[daily_test['Service Start Date'].isin(forecast_aligned['ds'])]

    # Replace 'actual_consumption' with the actual column name in your test set
    y_true = test_aligned['Consumption (KWH)']
    y_pred = forecast_aligned['yhat']

    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2 = r2_score(y_true, y_pred)
    
    # Plotting the actual and predicted values
    axes[i].plot(test_aligned['Service Start Date'], test_aligned['Consumption (KWH)'], label='Actual')
    axes[i].plot(forecast_aligned['ds'], forecast_aligned['yhat'], label='Predicted')

    # Set titles and labels
    axes[i].set_title(f"MAE: {mae},\t MAPE: {mape}%, \t R2: {r2}")
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Electric Consumption (KWH)')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels
    axes[i].legend()

    print(f"Mean Absolute Error: {mae}")
    print(f"Mean Absolute Percentage Error: {mape}%")
    print(f"R-squared metric: {r2}")
plt.tight_layout()  # Adjust the layout
plt.show()
```

## Tuning prophet model parameters
- growth set to logistic, and adding cap to the training dataframe before training the model
- growth set to flat

## Growth - Flat

```python
# Assuming 'train_365' and 'test_365' are your dataframes
periods = [100, 200, 365]
fig, axes = plt.subplots(3, 1, figsize=(10, 15))  # Adjust figure size as needed

for i, period in enumerate(periods):
    # Prepare the data for Prophet
    prophet_data = daily_train.rename(columns={'Service Start Date': 'ds', 'Consumption (KWH)': 'y'})
    max_capacity = prophet_data['y'].max()*0.25
    prophet_data['cap'] = max_capacity
    
    # Initialize and configure the Prophet model
    prophet_model = Prophet(growth='flat',
                            n_changepoints=10,
                            changepoint_range=0.1,
                            changepoint_prior_scale=0.01,
                            seasonality_mode='additive',
                            yearly_seasonality=False,
                            weekly_seasonality=False,
                            daily_seasonality=False
                            ).add_seasonality(name='monthly', period=30, fourier_order=5
                            ).add_seasonality(name='quarterly',period=365*0.25,fourier_order=3
                            ).add_seasonality(name='yearly', period=365, fourier_order=14
                            ).add_seasonality(name='half yearly', period=365*0.5, fourier_order=3)
    
    prophet_model.fit(prophet_data)

    # Make future dataframe and predict
    future = prophet_model.make_future_dataframe(periods=period, include_history=False, freq='D')
    # Add 'cap' column to the future DataFrame
    future['cap'] = max_capacity
    forecast = prophet_model.predict(future)

    # Aligning the forecast with the test data
    forecast_aligned = forecast[forecast['ds'].isin(daily_test['Service Start Date'])]
    test_aligned = daily_test[daily_test['Service Start Date'].isin(forecast_aligned['ds'])]

    # Replace 'actual_consumption' with the actual column name in your test set
    y_true = test_aligned['Consumption (KWH)']
    y_pred = forecast_aligned['yhat']

    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    
    # Plotting the actual and predicted values
    axes[i].plot(test_aligned['Service Start Date'], test_aligned['Consumption (KWH)'], label='Actual')
    axes[i].plot(forecast_aligned['ds'], forecast_aligned['yhat'], label='Predicted')

    # Set titles and labels
    axes[i].set_title(f"MAE: {mae},\t MAPE: {mape}%")
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Electric Consumption (KWH)')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels
    axes[i].legend()

    print(f"Mean Absolute Error: {mae}")
    print(f"Mean Absolute Percentage Error: {mape}%")
plt.tight_layout()  # Adjust the layout
plt.show()

```

## Growth - Logistic

```python
# Assuming 'train_365' and 'test_365' are your dataframes
periods = [100, 200, 365]
fig, axes = plt.subplots(3, 1, figsize=(10, 15))  # Adjust figure size as needed

for i, period in enumerate(periods):
    # Prepare the data for Prophet
    prophet_data = daily_train.rename(columns={'Service Start Date': 'ds', 'Consumption (KWH)': 'y'})
    max_capacity = prophet_data['y'].max()*0.25
    prophet_data['cap'] = max_capacity
    
    # Initialize and configure the Prophet model
    prophet_model = Prophet(growth='logistic',
                            seasonality_mode='additive',
                            yearly_seasonality=False,
                            weekly_seasonality=False,
                            daily_seasonality=False
                            ).add_seasonality(name='monthly', period=30, fourier_order=22
                            ).add_seasonality(name='quarterly',period=365*0.25,fourier_order=3
                            ).add_seasonality(name='yearly', period=365, fourier_order=14
                            ).add_seasonality(name='half yearly', period=365*0.5, fourier_order=3)
    
    prophet_model.fit(prophet_data)

    # Make future dataframe and predict
    future = prophet_model.make_future_dataframe(periods=period, include_history=False, freq='D')
    # Add 'cap' column to the future DataFrame
    future['cap'] = max_capacity
    forecast = prophet_model.predict(future)

    # Aligning the forecast with the test data
    forecast_aligned = forecast[forecast['ds'].isin(daily_test['Service Start Date'])]
    test_aligned = daily_test[daily_test['Service Start Date'].isin(forecast_aligned['ds'])]

    # Replace 'actual_consumption' with the actual column name in your test set
    y_true = test_aligned['Consumption (KWH)']
    y_pred = forecast_aligned['yhat']

    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    
    # Plotting the actual and predicted values
    axes[i].plot(test_aligned['Service Start Date'], test_aligned['Consumption (KWH)'], label='Actual')
    axes[i].plot(forecast_aligned['ds'], forecast_aligned['yhat'], label='Predicted')

    # Set titles and labels
    axes[i].set_title(f"MAE: {mae},\t MAPE: {mape}%")
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Electric Consumption (KWH)')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels
    axes[i].legend()

    print(f"Mean Absolute Error: {mae}")
    print(f"Mean Absolute Percentage Error: {mape}%")
plt.tight_layout()  # Adjust the layout
plt.show()
```
